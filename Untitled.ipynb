```python
class Stack:
    
    def __init__(self):
        self.stack = []
        
    def add(self,newdata):
        if newdata not in self.stack:
            self.stack.append(newdata)
            return True
        else:
            return False
            
    def remove(self):
        if len(self.stack)<=0:
            return("No element in stack")
        else:
            return self.stack.pop()
            
    def printing(self):
        return self.stack

lst = Stack()
lst.add("Mon")
lst.remove()
lst.printing()
```

```python
class Queue:
    
    def __init__(self):
        self.queue = []
        
    def add(self,newdata):
        if newdata not in self.queue:
            self.queue.insert(0,newdata)
            return True
        return False
    
    def remove(self):
        if len(self.queue)>0:
            return self.queue.pop()
        return("No element in queue")
    
    def printing(self):
        return self.queue
    
TheQueue = Queue()
TheQueue.add("Mon")
TheQueue.add("Tue")
TheQueue.printing()
TheQueue.remove()
TheQueue.printing()
```

```python
import random

def play():
    user = input("'r' for rock,'p' for paper,'s' for scissors.")
    computer = random.choice(['r','p','s'])
    
    
    if user == computer:
        return "It's a Tie"
    


    def is_win(player,opponent):
        if (player =='r' and opponent == 's') or (player =='s' and opponent=='p') or(player =='p' and opponent == 'r'):
            return True 
        
    if is_win(user,computer):
        return 'You Won!'
    return 'You Lost!'
print(play())        
```
